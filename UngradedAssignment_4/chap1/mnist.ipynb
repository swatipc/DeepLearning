```python
import torch 
import torch.nn as nn
import torch.nn.functional as F
import torch.optim as optim
from torchvision import datasets, transforms
```

```python
torch.manual_seed(4242)


```

```python
train_loader = torch.utils.data.DataLoader(
    datasets.MNIST('/content/drive/My Drive/Colab Notebooks/Deep Learning/DL_ungraded_assignment4/DL_ungraded_assignment4/chap1/Input/mnist/', train =True, download=True,
                  transform=transforms.Compose([
                      transforms.ToTensor(),
                      transforms.Normalize((0.1307,),(0.3081))
                     ])),
    batch_size=64, shuffle=True)

```

```python
class Net(nn.Module):
    def __init__(self):
        super(Net, self).__init__()
        self.conv1 = nn.Conv2d(1,10,kernel_size=5)
        self.conv2 = nn.Conv2d(10,20,kernel_size=5)
        self.conv2_drop = nn.Dropout2d()
        self.fc1= nn.Linear(320,50)
        self.fc2 =nn.Linear(50,10)
    
    def forward(self, a):
        a= F.relu(F.max_pool2d(self.conv1(a),2))
        a= F.relu(F.max_pool2d(self.conv2_drop(self.conv2(a)), 2))
        a = a.view(-1,320)
        a = F.relu(self.fc1(a))
        a = F.dropout(a, training=self.training)
        a=self.fc2(a)
        return F.log_softmax(a, dim=1)
    


        
```

```python
my_model = Net()
```

```python
optimizer = optim.SGD(my_model.parameters(), lr=0.01, momentum=0.5)
```

```python
for epoch in range(10):
    for batch_idx, (data, target) in enumerate(train_loader):
        optimizer.zero_grad()
        output = my_model(data)
        loss = F.nll_loss(output, target)
        loss.backward()
        optimizer.step()
print('Current_loss: {}'.format(float(loss)))
```
