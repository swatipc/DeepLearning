```
#Define input and output layer
op_layer = data['Class']
ip_layer = data.drop(columns=['Class'])
```

```
#Split the dataset into Test and Train
x_train,x_test,y_train,y_test = train_test_split(ip_layer,op_layer,test_size=0.3,random_state=0)
x_train.shape
```

```
# Boolean Crossentrpoy algorithm

def boolean_CE_Algo(X_train,X_test,y_train,y_test):
    l = Sequential()
    l.add(Dense(300, activation="relu",input_shape=(10,)))
    l.add(Dropout(0.2))
    l.add(Dense(10, activation="relu"))
    l.add(Dropout(0.1))
    l.add(Dense(10, activation="relu"))
    l.add(Dense(1,activation='sigmoid'))
    l.compile(loss='binary_crossentropy', optimizer=sgd(), metrics=['accuracy'])
    l.fit(x_train, y_train, epochs=10)
    res = l.evaluate(x_test, y_test)
    return res
```
