```
train_source = image_dir + 'Train/'
test_source = image_dir + 'Test/'
```

```
test_source
```

```
train_dataGenerator = ImageDataGenerator(rescale=1./255, 
                                   shear_range=0.2, 
                                   zoom_range=0.2, 
                                   rotation_range=20, 
                                   width_shift_range=0.2,
                                   height_shift_range=0.2, 
                                   horizontal_flip=True, 
                                   fill_mode='nearest', 
                                   validation_split=0.2) # set validation split
```

```
test_dataGenerator = ImageDataGenerator(rescale = 1./255)
```

```
# Default Values
#img_height = 150
#img_width = 150
batch_size = 32
```

```
train_generator = train_dataGenerator.flow_from_directory(train_source,
                                                    target_size=(64,64), 
                                                    batch_size=batch_size, 
                                                    class_mode=None,
                                                    subset='training') # set as training data

```

```
validation_generator = test_dataGenerator.flow_from_directory(test_source, 
                                                         target_size=(64,64), 
                                                         batch_size=batch_size, 
                                                         class_mode=None, 
                                                         subset='validation') # set as validation data
```

```
# defining global variable path
# Location of my dataset on My Google Drive
image_path = "/content/drive/My Drive/data/Fruits"
import tensorflow as tf

def loadImages(path):
    '''Put files into lists and return them as one list with all images 
     in the folder'''
    image_files = sorted([os.path.join(path, 'apple', file)
                          for file in os.listdir(path + "/Apple/")
                          if file.endswith('.png')])
    return image_files

```
