```
# Defining the input and output layer of Keras Model
op_layer = data['Glass_Type']
ip_layer = data.drop(columns=['Glass_Type'])
```

```
# Encode the lables present in output layer

encode = LabelEncoder()
encode.fit(op_layer)
encode_op = encode.transform(op_layer)
data_op = np_utils.to_categorical(encode_op)
print(data_op)
```

```
# Split the data into test and train
x_train,x_test,y_train,y_test = train_test_split(ip_layer,data_op,test_size=.3,random_state=0)
```

```
# Define a NN model for multiclass classification

def multiClass_algo(x_train,x_test,y_train,y_test):
    l = Sequential()
    l.add(Dense(10, activation="relu",input_shape=(9,)))
    l.add(Dropout(0.2))
    l.add(Dense(8, activation="relu"))
    l.add(Dropout(0.1))
    l.add(Dense(6,activation='softmax'))
    l.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy'])
    l.fit(x_train, y_train, epochs=500)
    res = l.evaluate(x_test, y_test)
    return res
```

```
# Run the model
multiClass_algo(x_train,x_test,y_train,y_test)
```
